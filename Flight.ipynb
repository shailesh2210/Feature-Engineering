```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df["Date_of_Journey"].str.split("/").str[0]
```

```python
df["Day"] = df["Date_of_Journey"].str.split("/").str[0]
df["Month"] = df["Date_of_Journey"].str.split("/").str[1]
df["Year"] = df["Date_of_Journey"].str.split("/").str[2]
```

```python
df.head()
```

```python
df["Dep_hour"] = df["Dep_Time"].str.split(":").str[0]
df["Dep_min"] = df["Dep_Time"].str.split(":").str[1]
```

```python
df.head()
```

```python
# df['Arrival_Time']=df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
df["Arrival_Time"] = df["Arrival_Time"].apply(lambda x: x.split(" ")[0])
```

```python
df["Arrival_hour"] = df["Arrival_Time"].str.split(":").str[0]
df["Arrival_min"] = df["Arrival_Time"].str.split(":").str[1]
```

```python
df.head()
```

```python
df["Total_Stops"].unique()
```

```python
df["Total_Stops"] = df["Total_Stops"].map({'non-stop': 0,'1 stop':1, '2 stops':2 ,'3 stops':3,"'4 stops'":4,"nan":1  })
```

```python
df.head()
```

```python
df.drop(["Date_of_Journey","Route","Dep_Time","Dep_Time"], axis=1, inplace = True)
```

```python
df.head()
```

```python
df.drop(["Arrival_Time"], axis=1, inplace= True)
```

```python
df.head()
```

```python
df["Duration_hour"] = df["Duration"].str.split("h").str[0]
```

```python
df.head()
```

```python
df.drop("Duration", axis=1, inplace =True)
```
